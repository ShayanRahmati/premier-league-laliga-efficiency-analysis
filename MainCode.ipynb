```python
import pandas as pd
```

```python
df = pd.read_csv("project2.csv")
```

```python
df["Date"] = pd.to_datetime(df['Date'])
```

```python
def get_season(date):
    if date.month>=8:
        return str(date.year) + '-' + str(date.year+1)
    else:
        return str(date.year-1) + '-' + str(date.year)
```

```python
df['Season'] = df ['Date'].apply(get_season)
```

```python
df = df[df['Season'] != '2023-2024']
```

```python
df[['home_goals', 'away_goals']] = df['Score'].str.split('–', expand=True).astype(float)
```

```python
home_df = df[["Season", "Home", "home_goals", "xG"]].copy()
```

```python
home_df.columns = ["Season", "Team", "Goals", "xG"]
```

```python
home_df["Venue"] = "Home"
```

```python
home_df["Efficiency"] = home_df["Goals"] - home_df["xG"]
```

```python
away_df = df[["Season", "Away", "away_goals", "xG.1"]].copy()
```

```python
away_df.columns = ["Season", "Team", "Goals", "xG"]
```

```python
away_df["Venue"] = "Away"
```

```python
away_df["Efficiency"] = away_df["Goals"] - away_df["xG"]
```

```python
teams_df = pd.concat([home_df, away_df])
```

```python
teams_df = teams_df.sort_values(by=["Season", "Team"])
```

```python
teams_df = teams_df.reset_index(drop=True)
```

```python
teams_df.to_csv("final_data.csv", index=False)
```
